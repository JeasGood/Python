```
import numpy as np
import matplotlib.pyplot as plt
import cv2

im_1 = cv2.imread("Lab3_half100.png", cv2.IMREAD_GRAYSCALE)
im_2 = cv2.imread("Lab3_chess100.png", cv2.IMREAD_GRAYSCALE)

im_h = cv2.cvtColor(im_1, cv2.COLOR_BGR2RGB)
im_ch = cv2.cvtColor(im_2, cv2.COLOR_BGR2RGB)






plt.figure(figsize=(20,10))

plt.subplot(2,2,1)
plt.imshow(im_h, cmap = 'gray')

plt.subplot(2,2,2)
h_h = cv2.calcHist([im_h], [0], None, [256], [0, 256])
plt.plot(h_h); plt.xlim([0, 255]); plt.grid(); plt.plot();

plt.show()

plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plt.imshow(im_ch,cmap = 'gray')

plt.subplot(2,2,2)
c_h = cv2.calcHist([im_ch], [0], None, [256], [0, 256])
plt.plot(c_h); plt.xlim([0, 255]);  plt.grid(); plt.plot();


plt.show()
```

```
plt.figure(figsize=(20,10))
plt.subplot(231)
R = np.zeros(im_h.shape, dtype='uint8')
R[:, :, 0] = im_h[:, :, 0]
plt.imshow(R)

plt.subplot(232)
G = np.zeros(im_h.shape, dtype='uint8')
G[:, :, 1] = im_h[:, :, 1]
plt.imshow(G)

plt.subplot(233)
B = np.zeros(im_h.shape, dtype='uint8')
B[:, :, 2] = im_h[:, :, 2]
plt.imshow(B)
plt.show()

plt.figure(figsize=(20,10))
plt.subplot(231)
RC = np.zeros(im_ch.shape, dtype='uint8')
RC[:, :, 0] = im_ch[:, :, 0]
plt.imshow(RC)

plt.subplot(232)
GC = np.zeros(im_ch.shape, dtype='uint8')
GC[:, :, 1] = im_ch[:, :, 1]
plt.imshow(GC)

plt.subplot(233)
BC = np.zeros(im_ch.shape, dtype='uint8')
BC[:, :, 2] = im_ch[:, :, 2]
plt.imshow(BC)
plt.show()
```

```
plt.figure(figsize=(20,10))
plt.subplot(241)
D=R[:, :, 0] + RC[:, :, 0]
E=G[:, :, 1] + GC[:, :, 1]
F=B[:, :, 2] + BC[:, :, 2]
img_merge = cv2.merge((D, E, F ))
plt.imshow(img_merge)
plt.subplot(242)
plt.imshow(D)
plt.subplot(243)
plt.imshow(E)
plt.subplot(244)
plt.imshow(F)
plt.imshow(F)
plt.show()
```
